```python
import numpy as np
x= np.array([10,30,2, 59, 90, 80])
y= np.array([90, 34, 56, 78,23, 77])
```

```python
import numpy as np

def Pearson_correlation(x, y):
    if len(x) == len(y):
        sum_xy = np.sum((x - x.mean()) * (y - y.mean()))
        sum_x_squared = np.sum((x - x.mean()) ** 2)
        sum_y_squared = np.sum((y - y.mean()) ** 2)

        corr = sum_xy / np.sqrt(sum_x_squared * sum_y_squared)
        return corr
    else:
        return "Length of X and Y must be equal"

print(Pearson_correlation(x, y))
print(Pearson_correlation(x, x))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_heatmap = {
    'A': [10, 20, 30, 40],
    'B': [50, 60, 70, 80],
    'C': [90, 100, 110, 120],
    'D': [130, 140, 150, 160]
}

df_heatmap = pd.DataFrame(data_heatmap)

# Generate heatmap
sns.heatmap(df_heatmap, annot=True, fmt="d", cmap="YlGnBu")

plt.show()
```
