```python
import pandas as pd
import numpy as np
import scipy as sp
```

```python
data_string = '10,20,30,40,50'
data = np.array([int(x) for x in data_string.split(',')])
mean_data =  data.mean()
print("mean", mean_data)
```

```python
outlier_data = np.array([10,200,30,40,50])
mean_outlier = outlier_data.mean()
print("outlier mean data:", mean_outlier)
```

```python
from scipy import stats
zscores = np.abs(stats.zscore(outlier_data))
outliers = outlier_data[(zscores > 3)]
```

```python
q1 = np.percentile(outlier_data, 25)
q3 = np.percentile(outlier_data, 75)
IQR = q3-q1
outliers = outlier_data[(outlier_data < (q1 - 1.5 * IQR)) | (outlier_data > (q3 + 1.5 * IQR))]
```
